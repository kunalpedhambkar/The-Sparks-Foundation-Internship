### Linear Regression:
*Linear regression attempts to model the relationship between two variables by fitting a linear equation to observed data. One variable is considered to be an explanatory variable, and the other is considered to be a dependent variable. For example, a modeler might want to relate the weights of individuals to their heights using a linear regression model.*

### Equation for Linear Regression:
A linear regression line has an equation of the form **Y = a + bX** where,

X -> the independent variable 

Y -> the dependent variable.

b -> slope of the line

a -> the intercept (the value of y when x = 0).

### Importing the Required Libraries: 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline 
```

### Importing & Reading the Data:

```python
df= pd.read_csv("https://raw.githubusercontent.com/AdiPersonalWorks/Random/master/student_scores%20-%20student_scores.csv")
print(df.head(10))
print("")
print("The Data has been imported!")
```

### Plot the Dataset:

```python
df.plot(x='Hours', y='Scores', style='o')  
plt.title('Hours vs Scores')  
plt.xlabel('Hours Studied')  
plt.ylabel('Score Obtained')  
plt.show()
```

### Preparing the Data:

```python
from sklearn.model_selection import train_test_split
X= df.iloc[:, :-1].values
y= df.iloc[:, 1].values
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

### Training the Dataset:

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()  
regressor.fit(X_train, y_train) 
print("The Training of the Linear Model is Complete!")
```

### Predicting the Result:

```python
y_pred = regressor.predict(X_test)
```

### Plotting the Regression Line on Training Data:

```python
plt.scatter(X_train, y_train, color = 'red')
plt.plot(X_train, regressor.predict(X_train), color = 'blue')
plt.title('Hours vs Scores')
plt.xlabel('Hours Studied')
plt.ylabel('Score Obtained')
plt.show()
```

### Plotting the Regression Line on Test Data:

```python
plt.scatter(X_test, y_test, color = 'red')
plt.plot(X_train, regressor.predict(X_train), color = 'blue')
plt.title('Hours vs Scores')
plt.xlabel('Hours Studied')
plt.ylabel('Score Obtained')
plt.show()
```

### Comparing Actual & Predicted Values:

```python
comp = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})  
print(comp)
```

### Testing the Model with Asked Value:

```python
hours= [[9.25, ]]
ans = regressor.predict(hours)
print("No of Hours = {}".format(hours[0][0]))
print("Predicted Score = {}".format(ans[0]))
```

### Evaluating the Model:

```python
from sklearn import metrics  
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_pred)) 
```
